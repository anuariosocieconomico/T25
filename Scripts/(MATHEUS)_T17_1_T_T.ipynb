```python
#PEGANDO OS DADOS DE SERGIPE DO GRAFICO T16.2

T17p1_SE = sidrapy.get_table(table_code="7143",variable="10274/C872/47821,47823,47825,47826/C11797/95298,47820", territorial_level="3", ibge_territorial_code="28", period="all")
T17p1_SE = T17p1_SE.iloc[1:, [8,6,12,14,4]]
T17p1_SE.columns = ['ANO','REGIÃO','REDE DE ENSINO','CURSO FREQUENTADO','DISTRIBUIÇÃO DAS PESSOAS QUE FREQUENTAM ESTABELECIMENTO DE ENSINO (%)']
T17p1_SE

#PEGANDO OS DADOS DO NORDESTE DO GRAFICO T16.2

T17p1_NE = sidrapy.get_table(table_code="7143",variable="10274/C872/47821,47823,47825,47826/C11797/95298,47820", territorial_level="2", ibge_territorial_code="2", period="all")
T17p1_NE = T17p1_NE.iloc[1:, [8,6,12,14,4]]
T17p1_NE.columns = ['ANO','REGIÃO','REDE DE ENSINO','CURSO FREQUENTADO','DISTRIBUIÇÃO DAS PESSOAS QUE FREQUENTAM ESTABELECIMENTO DE ENSINO (%)']
T17p1_NE

#PEGANDO OS DADOS DO BRASIL DO GRAFICO T16.2

T17p1_BR = sidrapy.get_table(table_code="7143",variable="10274/C872/47821,47823,47825,47826/C11797/95298,47820", territorial_level="1", ibge_territorial_code="1", period="all")
T17p1_BR = T17p1_BR.iloc[1:, [8,6,12,14,4]]
T17p1_BR.columns = ['ANO','REGIÃO','REDE DE ENSINO','CURSO FREQUENTADO','DISTRIBUIÇÃO DAS PESSOAS QUE FREQUENTAM ESTABELECIMENTO DE ENSINO (%)']
T17p1_BR

# JUNTANDO TODAS AS TABELA

T17P1 = pd.concat([T17p1_SE,T17p1_NE,T17p1_BR])

#CONFIGURANDO OS TIPOS DE ARQUIVO DAS VARIÁVEIS
T17P1['ANO'] = T17P1['ANO'].astype(np.int64)
T17P1['DISTRIBUIÇÃO DAS PESSOAS QUE FREQUENTAM ESTABELECIMENTO DE ENSINO (%)'] = T17P1['DISTRIBUIÇÃO DAS PESSOAS QUE FREQUENTAM ESTABELECIMENTO DE ENSINO (%)'].astype(float)

#EXPORTANDO (CSV E EXCEL)
T17P1.to_csv('T17_1.csv', sep=';', index=False, encoding='latin1', decimal=',')
T17P1.to_excel('T17_1.xlsx', index=False,)

# VISUALIZANDO
T17P1
```
