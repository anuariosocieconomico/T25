```python
#PEGANDO OS DADOS DE SERGIPE DO GRAFICO G16.1
g16p1_SE = sidrapy.get_table(table_code="6578",variable="10163", territorial_level="3", ibge_territorial_code="28", period="all")
g16p1_SE = g16p1_SE.iloc[1:, [7,6,4]]
g16p1_SE.columns = ['ANO','REGIÃO','NÚMERO MÉDIO DE PESSOAS POR DOMICÍLIO']
g16p1_SE

#PEGANDO OS DADOS DO NORDESTE DO GRAFICO G16.1
g16p1_NE = sidrapy.get_table(table_code="6578",variable="10163", territorial_level="2", ibge_territorial_code="2", period="all")
g16p1_NE = g16p1_NE.iloc[1:, [7,6,4]]
g16p1_NE.columns = ['ANO','REGIÃO','NÚMERO MÉDIO DE PESSOAS POR DOMICÍLIO']
g16p1_NE

#PEGANDO OS DADOS DO BRASIL DO GRAFICO G16.1
g16p1_BR = sidrapy.get_table(table_code="6578",variable="10163", territorial_level="1", ibge_territorial_code="1", period="all")
g16p1_BR = g16p1_BR.iloc[1:, [7,6,4]]
g16p1_BR.columns = ['ANO','REGIÃO','NÚMERO MÉDIO DE PESSOAS POR DOMICÍLIO']
g16p1_BR

# JUNTANDO TODAS AS TABELA

G16P1 = pd.concat([g16p1_BR, g16p1_NE,g16p1_SE,])

#CONFIGURANDO OS TIPOS DE ARQUIVO DAS VARIÁVEIS
G16P1['ANO'] = G16P1['ANO'].astype(np.int64)
G16P1['NÚMERO MÉDIO DE PESSOAS POR DOMICÍLIO'] = G16P1['NÚMERO MÉDIO DE PESSOAS POR DOMICÍLIO'].astype(float)

#EXPORTANDO (CSV E EXCEL)
G16P1.to_csv('G16_1.csv', sep=';', index=False, encoding='latin1', decimal=',')
G16P1.to_excel('G16_1.xlsx', index=False,)

# VISUALIZANDO
G16P1
```
