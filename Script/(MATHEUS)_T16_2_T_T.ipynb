```python
#PEGANDO OS DADOS DE SERGIPE DO GRAFICO T16.2

T16p2_SE = sidrapy.get_table(table_code="6677",variable="10250/C845/47552,47553,47554,47555,47556", territorial_level="3", ibge_territorial_code="28", period="all")
T16p2_SE = T16p2_SE.iloc[1:, [8,6,12,4]]
T16p2_SE.columns = ['ANO','REGIÃO','POSSE DE BENS','PERCENTUAL DE DOMICÍLIOS']
T16p2_SE

#PEGANDO OS DADOS DO NORDESTE DO GRAFICO T16.2

T16p2_NE = sidrapy.get_table(table_code="6677",variable="10250/C845/47552,47553,47554,47555,47556", territorial_level="2", ibge_territorial_code="2", period="all")
T16p2_NE = T16p2_NE.iloc[1:, [8,6,12,4]]
T16p2_NE.columns = ['ANO','REGIÃO','POSSE DE BENS','PERCENTUAL DE DOMICÍLIOS']
T16p2_NE

#PEGANDO OS DADOS DO BRASIL DO GRAFICO T16.2

T16p2_BR = sidrapy.get_table(table_code="6677",variable="10250/C845/47552,47553,47554,47555,47556", territorial_level="1", ibge_territorial_code="1", period="all")
T16p2_BR = T16p2_BR.iloc[1:, [8,6,12,4]]
T16p2_BR.columns = ['ANO','REGIÃO','POSSE DE BENS','PERCENTUAL DE DOMICÍLIOS']
T16p2_BR

# JUNTANDO TODAS AS TABELA

T16P2 = pd.concat([T16p2_SE, T16p2_NE,T16p2_BR,])

#CONFIGURANDO OS TIPOS DE ARQUIVO DAS VARIÁVEIS
T16P2['ANO'] = T16P2['ANO'].astype(np.int64)
T16P2['PERCENTUAL DE DOMICÍLIOS'] = T16P2['PERCENTUAL DE DOMICÍLIOS'].astype(float)

#EXPORTANDO (CSV E EXCEL)
T16P2.to_csv('T16_2.csv', sep=';', index=False, encoding='latin1', decimal=',')
T16P2.to_excel('T16_2.xlsx', index=False,)

# VISUALIZANDO
T16P2
```
